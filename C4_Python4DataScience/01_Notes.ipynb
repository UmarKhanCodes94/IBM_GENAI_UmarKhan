```python
#exception handling 

# try: , except error: , else: , finally:   
''' 
Step-by-Step Explanation:

    try Block
        The code that might raise an exception is placed inside try.
        If an exception occurs, execution jumps to the except block.

    except Block(s)
        Catches and handles specific exceptions.
        except ZeroDivisionError: handles division by zero.
        except TypeError: handles invalid data types.

    else Block (Optional)
        Runs only if the try block succeeds (i.e., no exception occurs).
        Here, it prints the result.

    finally Block (Always Executes)
        Runs whether an exception occurs or not.
        Useful for cleanup tasks (e.g., closing files, releasing resources).

'''
```

```python
#objects and classes 

# all data types are basically objects 
#every object has 
    # a type 
    # an internal data representation (blueprint)
    # a set of procedures to interact with the object (methods) // list.sort() changes the object // list.reverse()
# an object is an instance of a particuler type 

#building class 
    # data attributes
    # methods 

class circle(object):
    
    def __init__(self, radius,color): #class constructor / special function with parameters (self -> refers to newly created instance)
    
        self.radius=radius 
        self.color=color

    def add_radius(self,r):
        self.radius=self.radius+r

class rectangle(object):

    def __init__ (self, height, width, color):
    
        self.height = height 
        self.width = width 
        self.color = color 

red_circle = circle(10,'red')
green_rectangle = rectangle(10,10,'green')
```

```python
# we can use python BeautifulSoup for HTML parsing 
from bs4 import BeautifulSoup
```

```python
# Specify the URL of the webpage you want to scrape
url = 'https://en.wikipedia.org/wiki/IBM'
# Send an HTTP GET request to the webpage
response = requests.get(url)
# Store the HTML content in a variable
html_content = response.text
# Create a BeautifulSoup object to parse the HTML
soup = BeautifulSoup(html_content, 'html.parser')
# Display a snippet of the HTML content
print(html_content[:500])
```

```python
# Find all <a> tags (anchor tags) in the HTML
links = soup.find_all('a')
# Iterate through the list of links and print their text
for link in links:
    print(link.text)
```
